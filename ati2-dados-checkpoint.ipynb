```python
df = pd.read_excel(r"C:\Users\DYENIFER\Downloads\default_of_credit_card_clients__courseware_version_1_21_19.xls")
```

```python
caracteristicas_fatura = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']

caracteristicas_pagamento = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
```

```python
df[caracteristicas_fatura].describe()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, coluna in enumerate(caracteristicas_fatura):
    ax = axes[i//3, i%3]
    ax.hist(df[coluna].dropna(), bins=20, color='skyblue', edgecolor='black')
    ax.set_title(coluna)
    ax.set_xlabel('Valor')
    ax.set_ylabel('Frequncia')

plt.tight_layout()
plt.show()
```

```python
df[caracteristicas_pagamento].describe()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, coluna in enumerate(caracteristicas_pagamento):
    ax = axes[i//3, i%3]
    ax.hist(df[coluna].dropna(), bins=20, color='salmon', edgecolor='black')
    ax.set_title(coluna)
    ax.set_xlabel('Valor')
    ax.set_ylabel('Frequncia')
    ax.tick_params(axis='x', rotation=45)  # Rotacionando os rtulos do eixo X

plt.tight_layout()
plt.show()
```

```python
pagamentos_zeros = (df[caracteristicas_pagamento] == 0).sum()
print('Nmero de pagamentos iguais a 0 por coluna:')
print(pagamentos_zeros)
```

```python
import numpy as np

df_log_pagamentos = df[caracteristicas_pagamento].applymap(lambda x: np.log10(x) if x > 0 else np.nan)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, coluna in enumerate(caracteristicas_pagamento):
    ax = axes[i//3, i%3]
    ax.hist(df_log_pagamentos[coluna].dropna(), bins=20, color='lightgreen', edgecolor='black')
    ax.set_title(f'Log({coluna})')
    ax.set_xlabel('Log do Valor')
    ax.set_ylabel('Frequncia')

plt.tight_layout()
plt.show()
```
